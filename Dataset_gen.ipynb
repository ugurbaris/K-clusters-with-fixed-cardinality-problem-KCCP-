```python
def dataset_gen(file,N):
    import pandas as pd
    import numpy as np
    import random

    data = pd.read_csv('C:\\Users\\ugur_\\dataset_softw_arc\\'+file+'.txt',delimiter= '\s+', header=None)
    data.columns = ["n1", "n2", "w"]

    dfd = data.copy()

    for idx,weight in enumerate (data["w"]):
        weight = round(random.uniform(0, 1),2)
        print(weight)
        dfd.loc[idx, 'w'] = weight

    nodes=(set(data["n1"])).union(set(data["n2"]))

    graf= random.sample(nodes,N)

    m=len(graf)
    size = (m,m)
    matrix = np.zeros(size)

    dicto = {}
    for idx,each in enumerate(graf):
        dicto[each] = idx


    for node1,node2,w in zip(dfd["n1"],dfd["n2"],dfd["w"]):
        if (node1 in graf and node2 in graf) or (node2 in graf and node1 in graf):
            print(node1,node2,w)
            matrix[dicto[node1],dicto[node2]] = w
            matrix[dicto[node2],dicto[node1]] = w

    df2 = pd.DataFrame(data=matrix.astype(float))
    df2.to_csv('C:\\Users\\ugur_\\Documents\\MatrixForm_'+file+'.txt', sep=',', header=False, float_format='%.2f', index=False)

    file1 = open('C:\\Users\\ugur_\\Documents\\MatrixForm_'+file+'.txt',"r+")
    readcontent = file1.read()

    file1.seek(0, 0)
    file1.write("SIM : [ " + "\n" )

    file1.write(readcontent)

    file1.write("]")
    file1.close()


```

```python
file = "PB1_d025"
N=40
dataset_gen(file,N)
```
